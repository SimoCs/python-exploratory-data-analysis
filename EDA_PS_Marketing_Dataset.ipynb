# Analyzing the PS Marketing Dataset

## Setup:

```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import numpy as np
```

##  Exploratory Data Analysis (EDA):

To conduct a full Exploratory Data Analysis (EDA) on your dataset, I'll start by loading the data to understand its structure, content, and the types of data it contains. This will involve:
*   Checking the first few rows to get a feel for the data.
*   Understanding the data types of each column.
*   Checking for missing values.
*   Summarizing the statistical properties of the numerical data.
*   Exploring the distribution of categorical variables.
*   Visualizing the data to uncover trends and patterns.


Let's begin by loading the data and taking a look at the first few rows:

```python
# Load the dataset
marketing_data = pd.read_csv("Marketing - recordid_name_gender_age_location_email_phone_product_category_amount.csv")

# Display the first few rows of the dataframe
marketing_data.head()
```

The dataset contains the following columns:
* **RecordID**: A unique identifier for each record.
* **Name**: The name of the customer.
* **Gender**: The gender of the customer.
* **Age**: The age of the customer.
* **Location**: The location of the customer.
* **Email**: The email address of the customer.
* **Phone**: The phone number of the customer.
* **Product**: The product purchased by the customer.
* **Category**: The category of the product.
* **Amount**: The amount spent on the purchase.

Next, I'll perform the following steps:
* Check the data types and look for any initial issues with the data.
* Check for any missing values.
* Summarize the dataset to get an overview of the numerical fields.
* Explore the distribution of the categorical variables.

```python
# Check the data types and look for any issues with the data
data_types = marketing_data.dtypes

data_types
```

```python
# Check for any missing values
missing_values = marketing_data.isnull().sum()

missing_values
```

```python
# Summarize the dataset for numerical fields
summary_statistics = marketing_data.describe()

summary_statistics
```

Here's a summary of the dataset's properties:
* The data types for each column are appropriate, with **RecordID** and **Age** being integers, **Amount** being a floating-point number, and the rest being objects (which typically denote strings in pandas).
* There are no missing values in any of the columns.
* There are **100** records in total.
* The **Age** column has values ranging from **24** to **45** with a mean age of approximately **31**.

Next steps in the EDA will include:
* Exploring the distribution of the **Age** and **Amount** columns.
* Examining the categorical variables (**Product** and **Category**) to understand the distribution of records across different classes.

Let's proceed with visualizing the distribution of the numerical data (**Age** and **Amount**) and then explore the categorical variables:

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Plot distribution of the Age column
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(marketing_data['Age'], bins=10, kde=True)
plt.title('Age Distribution')

# Plot distribution of the Amount column
plt.subplot(1, 2, 2)
sns.histplot(marketing_data['Amount'], bins=10, kde=True)
plt.title('Amount Spent Distribution')

plt.show()
```

The distribution plots provide the following insights:
*   The **Age** distribution is somewhat left-skewed, indicating that there are more younger customers in this dataset.
*   The **Amount** spent is right-skewed, which is common in spending data because there are typically more smaller transactions than large ones.

Next, let's explore the categorical variables (**Product** and **Category**) to understand the distribution of records across different classes.

For this, we'll create count plots for each categorical variable to visualize the number of records for each category.:

```python
# Plotting count plots for categorical variables

plt.figure(figsize=(18, 12))

# Plot distribution of the Product column
plt.subplot(2, 2, 3)
sns.countplot(x='Product', data=marketing_data)
plt.title('Product Distribution')
plt.xticks(rotation=90)

# Plot distribution of the Category column
plt.subplot(2, 2, 4)
sns.countplot(x='Category', data=marketing_data)
plt.title('Category Distribution')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

The count plots reveal the following about the categorical variables:


*   **Product Distribution**: The products are diverse. The visualization shows which products are more commonly purchased, though it's a bit cluttered due to the number of different products.
*   **Category Distribution**: Similar to products, the categories are diverse, with some categories being more popular than others.
